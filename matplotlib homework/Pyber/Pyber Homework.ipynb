```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
citydata=pd.read_csv('raw_data/city_data.csv')
citydata
```

```python
ridedata=pd.read_csv('raw_data/ride_data.csv')
ridedata
#list(ridedata)
```

```python
#Your objective is to build a Bubble Plot that showcases the relationship between four key variables:

#Average Fare ($) Per City
#Total Number of Rides Per City
#Total Number of Drivers Per City
#City Type (Urban, Suburban, Rural)

```

```python

citymean=ridedata.groupby(['city']).fare.agg(["mean","count"]).reset_index()
citymean
#list(citymean)
```

```python
citymerged = pd.merge(citymean, citydata, how="outer")
citymerged
```

```python


s=[10*n for n in citymerged['count']]

citymerged.plot(kind='scatter', x='count', y='mean',s = s,c='r',linewidth=1,edgecolor='black')


plt.show()
```

```python
urban=citymerged.loc[citymerged['type'] == "Urban"]
rural=citymerged.loc[citymerged['type'] == "Rural"]
suburban=citymerged.loc[citymerged['type'] == "Suburban"]

fig = plt.figure(figsize=(18,10))
ax = fig.add_subplot(111)

urban.plot(kind='scatter', x='count', y='mean',s = s,c='gold',ax=ax,label='Urban',alpha=0.7,linewidth=1,edgecolor='black')
rural.plot(kind='scatter', x='count', y='mean',s = s,c='lightskyblue',ax=ax,label='Rural',alpha=0.7,linewidth=1,edgecolor='black')
suburban.plot(kind='scatter', x='count', y='mean',s = s,c='coral',ax=ax,label='Suburban',alpha=0.7,linewidth=1,edgecolor='black')

plt.legend(loc='upper right')
plt.grid()

plt.title("Pyder Ride Sharing Data 2016")
plt.xlabel("Total Number of Ride Per City")
plt.ylabel("Average Fare ($)")

plt.show()
```

```python
#In addition, you will be expected to produce the following three pie charts:

#% of Total Fares by City Type
farebycity = citymerged.groupby(['type']).sum()
farebycity 
```

```python
farebycitytotal=farebycity['count'] * farebycity['mean']
farebycitytotal
```

```python
pie = plt.pie(farebycitytotal,autopct='%1.1f%%', startangle=90,)
labels = ['Rural','Suburban','Urban']
plt.legend( loc = 'best', labels=labels)
plt.title("% of Total Fares by City Type")
plt.show()
```

```python
#% of Total Drivers by City Type
farebycity.plot(kind='pie', y='driver_count',label='',autopct='%.2f',colors=colors,explode=explode, shadow=1)
plt.title(" % of Total Drivers by City Type")

plt.show()

```
