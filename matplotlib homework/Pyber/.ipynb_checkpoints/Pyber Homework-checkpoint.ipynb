```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
citydata=pd.read_csv('raw_data/city_data.csv')
citydata
```

```python
ridedata=pd.read_csv('raw_data/ride_data.csv')
ridedata
#list(ridedata)
```

```python
#Your objective is to build a Bubble Plot that showcases the relationship between four key variables:

#Average Fare ($) Per City
#Total Number of Rides Per City
#Total Number of Drivers Per City
#City Type (Urban, Suburban, Rural)

```

```python

citymean=ridedata.groupby(['city']).fare.agg(["mean","count"]).reset_index()
citymean
#list(citymean)
```

```python
citymerged = pd.merge(citymean, citydata, how="outer")
citymerged
```

```python


s=[2*n for n in citymerged['count']]

citymerged.plot(kind='scatter', x='count', y='mean',s = s,c='r')


plt.show()
```

```python
urban=citymerged.loc[citymerged['type'] == "Urban"]
rural=citymerged.loc[citymerged['type'] == "Rural"]
suburban=citymerged.loc[citymerged['type'] == "Suburban"]

fig = plt.figure()
ax = fig.add_subplot(111)

urban.plot(kind='scatter', x='count', y='mean',s = s,c='r',ax=ax,label='Urban')
rural.plot(kind='scatter', x='count', y='mean',s = s,c='b',ax=ax,label='Rural')
suburban.plot(kind='scatter', x='count', y='mean',s = s,c='g',ax=ax,label='Suburban')

plt.legend(loc='upper right')


plt.show()
```
