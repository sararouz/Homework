```python
import pandas as pd
import os
import json
import numpy as np

data1=pd.read_json('purchase_data.json')
data2=pd.read_json('purchase_data2.json')

data1
```

```python
data2
```

```python
datamerged = pd.merge(data1,data2,how="outer")
datamerged
```

```python
numberOfPlayer = datamerged["Gender"].count()
numberOfPlayer
```

```python
#Purchasing Analysis(Total)
#Number of Unique Items
#Average Purchase Price
#Total Number of Purchases
#Total Revenue

Items = datamerged.groupby(("Item Name")).mean()
Items
```

```python
NumberOfUniqueItems = Items.count()
NumberOfUniqueItems
```

```python
TotalRevenue=datamerged["Price"].sum()
TotalRevenue
```

```python
#Gender Demographics

#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed

CountOfgenders=datamerged.groupby(["Gender"]).count()
CountOfgenders=CountOfgenders.rename(columns={'Age':'Count','Item ID':'Percentage'})
CountOfgenders=CountOfgenders.drop(['Item Name','Price','SN'],axis=1)

CountOfgenders



```

```python
SumOfCounts = CountOfgenders["Percentage"].sum()
def Percentageapply(a):
    a=100*a/float(SumOfCounts)
    return a
CountOfgenders = CountOfgenders["Percentage"].apply(Percentageapply)
CountOfgenders
```

```python
#Age Demographics

#The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)

BinOfAge= [0,10,15,20,90]
GroupAge=["0-10","10-15","15-20","20-90"]
datamerged['Group of Age']=pd.cut(datamerged['Age'],BinOfAge, labels = GroupAge) 
datamerged
```

```python
#Purchase Count
PurchasedCount =datamerged.groupby(["Group of Age"]).size()
PurchasedCount
```

```python
#Average Purchase Price
datamerged.groupby(["Group of Age"]).mean()
```

```python
#Total Purchase Value
A = {'Group of Age':['count'], 'Price':['sum']}
SumByAge = datamerged.groupby(["Group of Age"]).agg(A)
SumByAge
```

```python
#Normalized Totals

MinPrice=SumByAge['Price'].min().values[0]
MaxPrice=SumByAge['Price'].max().values[0]
def NormalizedTotal(x):
    Answer=(x-MinPrice)/(MaxPrice-MinPrice)
    return Answer

SumByAge= SumByAge["Price"].applymap(NormalizedTotal)
SumByAge
```

```python
#Top Spenders

#Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
#SN
#Purchase Count
#Average Purchase Price
#Total Purchase Value
#Most Popular Items
f = {'Price':['sum','mean'], 'SN':['count'],'Item Name':[lambda x: ' '.join(x)]}
PurchasedNumberSN =datamerged.groupby(["SN"]).agg(f)
PurchasedNumberSN
```

```python

MostProfitableUser=PurchasedNumberSN.sort_values(by=[('Price', 'sum')], ascending=False)
MostProfitableUser.head(5)
```

```python
#Identify the 5 most popular items by purchase count, then list (in a table):
#Item ID
#Item Name
#Purchase Count
#Item Price
#Total Purchase Value
#Most Profitable Items
f = {'Price':['sum'], 'SN':['count']}
PurchasedPopular =datamerged.groupby(["Item Name","Item ID","Price"]).agg(f)
MostProfitableItems=PurchasedPopular.sort_values(by=[('Price', 'sum')], ascending=False)
MostProfitableItems.head(5)
```
